```python
# underfitting is when a model is not complex enough to be able to accurately predict the outcome when provided with data 
# overfitting is done when a model is too complex to the point where it is able to perfectly predict data it was trained on but
# not able to predict data it doesn't already have an example for
```

```python
# cross validation is when data is broken up into multiple sets of data with some set asside as training data and some set aside to
# be used as a test for the completed model.
```

```python
import pandas as pd
from sklearn import tree
from sklearn.metrics import ConfusionMatrixDisplay

dbdf = pd.read_csv("diabetes.csv")
```

```python
dbdf.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = dbdf.drop('Outcome', axis=1)
y = dbdf['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=42,)

sc = StandardScaler()
X_test = sc.fit_transform(X_test)
X_train = sc.fit_transform(X_train)

```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test)
```

```python
print(rf.feature_importances_, X.columns)
```

```python
import matplotlib.pyplot as plt

rf_tree = rf.estimators_[4]
tree.plot_tree(rf_tree)
plt.savefig('rf_tree.svg')
```
