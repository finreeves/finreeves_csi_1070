```python
import pandas as pd
import numpy as np
import math
import statistics
from scipy import stats
```

```python
belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
belize_df.head()
```

```python
new_df = belize_df[['neighbourhood','price','number_of_reviews','availability_365']].copy()
new_df.head()
```

```python
grouped_df = new_df.groupby('neighbourhood')
```

```python
price_df = grouped_df["price"].mean()
```

```python
review_df = grouped_df["number_of_reviews"].sum()
```

```python
availability_df = grouped_df["availability_365"].max()
```

```python
left_merge_df = pd.merge(price_df, review_df, on="neighbourhood", how='left')
left2_merge_df = pd.merge(left_merge_df, availability_df, on='neighbourhood',how='left')
left2_merge_df
```
