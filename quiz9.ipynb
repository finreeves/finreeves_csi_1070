```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
```

```python
dbdf = pd.read_csv("diabetes.csv")
```

```python
dbdf.head()
```

```python
from sklearn.model_selection import train_test_split
X = dbdf.drop('Outcome', axis=1)
y =dbdf['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                test_size=0.3,
                                                    random_state=42)
```

```python
model = tree.DecisionTreeClassifier(max_depth=7, max_features=6, random_state=42, min_impurity_decrease=0.01)
```

```python
model = model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
tree.plot_tree(model, feature_names=list(X.columns),
               filled=True, fontsize=6, rounded=True)

plt.show()
plt.savefig("tree.svg")
```
